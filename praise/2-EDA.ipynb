# Exploratory Data Analysis

```python
import pandas as pd
import panel as pn
pn.extension()
```

### Load the Data

```python
data = pd.read_csv('outputs/praise_data.csv')
total_hours = pd.read_csv('outputs/total_hours.csv')
receivers = pd.read_csv('outputs/receivers.csv')
quantifiers = pd.read_csv('outputs/quantifiers.csv')
```

```python
len(data)
```

### Validate Praise Dataset Against Total IH Dataset

```python
praise_dataset_aggregated = data.groupby('To')[['IH per Praise']].sum().sort_values('IH per Praise', ascending=False).reset_index()
# pn.Row(pn.Column("#### Praise Dataset Aggregated",praise_dataset_aggregated), pn.Column("#### Total Hours Dataset",total_hours[['Handle', 'Impact Hours']]))
```

### Compare Hours Received, Hours Deducted, and Hours Given

```python
receiving = receivers.groupby('To')[['IH per Praise']].sum().sort_values('IH per Praise', ascending=False).reset_index()
dished = receivers.groupby('From')[['IH per Praise']].sum().sort_values('IH per Praise', ascending=False).reset_index()
quantifying = quantifiers.groupby('To')[['IH per Praise']].sum().sort_values('IH per Praise', ascending=True).reset_index()
```

```python
pn.Row(
    pn.Column("#### Total Hours Given", dished), 
    pn.Column("#### Total Hours Received", receiving), 
    pn.Column("#### Total Hours Deducted", quantifying),
    pn.Column("#### Default Net Impact Hours", praise_dataset_aggregated),
)
```
