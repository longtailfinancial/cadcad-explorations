```python
default = pd.read_csv('outputs/YGG-AUGMENTATION.csv').replace('akrtws ', 'akrtws')
```

```python
default['Hatcher'] = default['Hatcher'].str.lower()
total_hours['Handle'] = total_hours['Handle'].str.lower()
default = default.merge(total_hours, left_on='Hatcher', right_on='Handle', how='right')
```

```python
default.loc[default['Handle'] == 'kristoferkristofer', 'Total Hours Deducted'] = -34.19
default['Total Hours Deducted'] = default['Total Hours Deducted'].fillna(0)
```

```python
default
```

```python
default = default.sort_values('Impact Hours', ascending=False)[['Handle', 'Total Hours Deducted', 'Impact Hours']]
```

```python
default['Paid Ratio'] = default['Total Hours Deducted'] / -default['Impact Hours']
default['Unpaid Bonus Factor'] = 1 - (default['Paid Ratio'] / default['Paid Ratio'].sum() * 10)

default
```

```python
praisemageddon['Name'] = praisemageddon['Name'].str.lower()
```

```python
combined = default.merge(praisemageddon, left_on='Handle', right_on='Name', how='left')
```

```python
combined.to_csv('outputs/ORIGINAL-IMPACT-HOURS.csv',index=False)
```

```python
combined['Name']  = combined['Name'].fillna(combined['Handle'])
```

```python
combined[combined.isna().any(axis=1)]
```

```python
combined['praisemaggedon'] = combined['Unpaid Bonus Factor'] * combined['IH held']
```

```python
combined['praisemaggedon'] = combined['praisemaggedon'] / combined['praisemaggedon'].sum()
combined.to_csv('outputs/PRAISEMAGGEDON.csv',index=False)
```

```python
combined
```

```python
combined['Initial Distribution'] = combined['Impact Hours'] / combined['Impact Hours'].sum()
```

```python
combined = combined.sort_values('praisemaggedon', ascending=False)[['Name', 'Initial Distribution', 'Unpaid Bonus Factor', 'praisemaggedon']].ffill()

combined['Final Distribution'] = combined['Initial Distribution'] * 0.52 + combined['praisemaggedon'] * 0.48
```

```python
magic_number = 9731.413
```

```python
combined['Final Hours'] = combined['Final Distribution'] * magic_number
```
