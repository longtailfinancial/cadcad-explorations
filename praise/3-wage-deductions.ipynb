```python
import pandas as pd
import panel as pn
import numpy as np
pn.extension()
```

```python
data = pd.read_csv('outputs/praise_data.csv')
total_hours = pd.read_csv('outputs/total_hours.csv')
receivers = pd.read_csv('outputs/receivers.csv')
quantifiers = pd.read_csv('outputs/quantifiers.csv')
received = receivers.groupby('To')[['IH per Praise']].sum().sort_values('IH per Praise', ascending=False).reset_index()
dished = receivers.groupby('From')[['IH per Praise']].sum().sort_values('IH per Praise', ascending=False).reset_index()
quantified = quantifiers.groupby('To')[['IH per Praise']].sum().sort_values('IH per Praise', ascending=True).reset_index()
net = data.groupby('To')[['IH per Praise']].sum().sort_values('IH per Praise', ascending=False).reset_index()
```

```python
df['Initial Distribution'] = df['Default Net Impact Hours'] / df['Default Net Impact Hours'].sum()
```
