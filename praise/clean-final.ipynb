As of this point we have a very clean and sanity checked dataset. The next steps are the following:
0. Apply a 'smoothing' on 'IH held' according to the ordering of 'Total Impact Hours'
1. Compute the 'Unpaid bonus factor' this is a measure of the payments received. It will be a 1.0 if no payments were received. 
3. Cap the 'Unpaid bonus factor' such that its minimum is 0.6
4. Multiply that factor by 'normalized' IH held (normalized = IH held / sum(IH held))
5. Normalize original distribution (Final Net Impact hours)
6. Combine for final distribution = 0.48 * normalized(praisemageddon) + 0.52 * normalized(Final Net Impact Hours) 

```python
combined['Paid Ratio'] = combined['Total Deductions'] / -combined['Total Impact Hours']
combined['Unpaid Bonus Factor'] = 1 - (combined['Paid Ratio'] / combined['Paid Ratio'].sum() * 10)
```
