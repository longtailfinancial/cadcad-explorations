```python
import pandas as pd
```

```python
quantifiers = pd.read_csv('outputs/quantifiers.csv')
```

```python
quantifying = quantifiers.groupby('To')[['IH per Praise']].sum().sort_values('IH per Praise', ascending=True).reset_index()
```

```python
quantifying
```

```python
results = pd.read_excel("data/TEC Praise Quantification.xlsx", sheet_name="Deducted IH Results (after audi", engine='openpyxl', header=1, usecols="A:B")
# results.to_csv('outputs/audit_results.csv',index=False)
```

```python
quantifying['To'] = quantifying['To'].str.lower()
results['Handle'] = results['Handle'].str.lower()
```

```python
merged = results.merge(quantifying, left_on='Handle', right_on='To', how='outer')
```

```python
merged[merged['Handle'].isna()].tail(40)
```

Since we are OK with the adjustments that do not appear in the final results, then we will drop them and move on.

```python
merged = merged[~merged['Handle'].isna()]
```

```python
merged
```

```python
merged['Diff'] = (merged['Total Deductions'] - merged['IH per Praise']).apply(lambda x: round(x,2))
```

```python
merged.sort_values('Diff')
```
