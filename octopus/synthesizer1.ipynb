```python
import param as pm
import panel as pn
import IPython.display as ipd
import numpy
pn.extension()
```

```python
class DrumBeat(pm.Parameterized):
    sr = 22050 # sample rate
    beat = pm.Number(-5, bounds=(-10,0), step=0.1)
    
    def make_beat(self):
        T = 2.0    # seconds
        t = numpy.linspace(0, T, int(T*sr), endpoint=False) # time variable
        x = numpy.exp(self.beat*t)*numpy.sin(2*numpy.pi*440*t)                # pure sine wave at 440 Hz
        return x
    
    def make_sound(self):
        beat = self.make_beat()
        return ipd.Audio(beat, rate=self.sr)
```

```python
drum = DrumBeat()
```

```python
pn.Column(drum, drum.make_sound)
```
