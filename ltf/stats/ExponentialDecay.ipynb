```python
import param as pm
import panel as pn
import numpy as np
import math
import pandas as pd
import hvplot.pandas
pn.extension()

class Decay(pm.Parameterized):
    l = pm.Number(0.5, bounds=(-10, 10), step=0.3)
    
    def f(self, x):
        return (math.exp(-self.l*x) - math.exp(-self.l)) / (1 - math.exp(-self.l))
    
    def decay(self):
        xs = np.linspace(0,1, int(1e3))
        y = np.vectorize(self.f)(xs)
        return pd.DataFrame([xs,y], index=['x','y']).T
    
    def view(self):
        return self.decay().hvplot(x='x',y='y', ylim=(0,1.1), xlim=(0,1.1))

d = Decay()

pn.Row(d, d.view)
```
